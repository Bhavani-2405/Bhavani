```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X= np.array([[4],[6],[8],[9],[13]])
Y=np.array([1,2,3,4,5])
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
Y_pred=model.predict(X)
```

```python
plt.scatter(X, Y, color='black')
plt.plot(X, Y_pred, color='pink')
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
print("Intercept (Î²0):", model.intercept_)
print("Coefficient (Î²1):", model.coef_)
```

```python
X = np.random.rand(100, 1) * 50  # 100 random numbers between 0 and 100
Y = 2 * X + np.random.randn(100, 1) * 10  # Linear relation with some noise
```

```python
plt.scatter(X, Y, color='pink')
plt.title("Generated Dataset of 100 Numbers")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
Y_pred = model.predict(X)
```

```python
plt.scatter(X, Y, color='pink')  # Actual data points
plt.plot(X, Y_pred, color='black')  # Regression line
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
print("Intercept (Î²0):", model.intercept_)
print("Coefficient (Î²1):", model.coef_)
```
